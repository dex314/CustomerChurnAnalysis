## Modeling

- Using a 3 Fold Stratified K Fold Cross Validation (Manual) With Xgboost


```python
from sklearn.model_selection import StratifiedKFold
```

```python
trainx, testx, trainy, testy = train_test_split(df0.drop('churn',axis=1), 
                                                y, test_size=0.15, random_state=123)
print(trainx.shape, testx.shape)
```

```python
folds = 3
skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state=123)
```

```python
val_preds=0
test_preds=0
feats=0

dtest = xgb.DMatrix(testx, label=testy)
for itrain, ival in skf.split(trainx, trainy):
    
    xt, yt = trainx.iloc[itrain], trainy.iloc[itrain]   
    xv, yv = trainx.iloc[ival], trainy.iloc[ival]
    print(xt.shape, xv.shape, yt.shape, yv.shape)
    dtrain = xgb.DMatrix(xt, label=yt)
    dval = xgb.DMatrix(xv, label=yv)

    params = {'max_depth':6,
              'eta': 0.02,
              'colsample_bytree':0.8,
              'alpha':2.0,
              'subsample':0.9,
              'reg_lambda':0.9,
              'reg_alpha':0.9,              
              'objective':'binary:logistic',
              'eval_metric':['auc','aucpr'],
              'scale_pos_weight':scale_pos_weight,
              'random_state':123}
    num_round=1000
    
    mod = xgb.train(params, dtrain, num_round, evals=[(dval,'valset')], verbose_eval=100, early_stopping_rounds=200)
    print("Val Set ========= ")
    val_preds = mod.predict(dval)
    print(classification_report(yv, np.round(val_preds).astype(int)))
    tp0 = mod.predict(dtest)
    test_preds += tp0
    gc.collect()
    
test_preds = test_preds / folds
```

### Feature Importances

```python
fig, ax=plt.subplots(1,1,figsize=(10,5))
xgb.plot_importance(mod, ax=ax,  max_num_features=20, importance_type='gain');
plt.title("Feature Importance From Last Model - Gain");

fig2, ax2=plt.subplots(1,1,figsize=(10,5))
xgb.plot_importance(mod, ax=ax2,  max_num_features=20, importance_type='weight', color='C1');
plt.title("Feature Importance From Last Model - Weight");
```

The month to month contract variable dominates the gain while variations of charges have the most weight.

For example, if we spend \\$X on trying to prevent those 123 customers from churning that were not going to churn, but we still save \\$Y from the 202 that were going to churn, then as long as \\$Y > \\$X the model might still be useful.

**In order to maximize our results, we can do two things:**

1. Threshold search for the minimum f1 score.
2. Deciling to focus only on those whose likelhood (lift) is far enough above the threshold.

```python
def threshold_search(y_true, y_proba):
    best_threshold = 0
    best_score = 0
    for threshold in tqdm_notebook([i * 0.01 for i in range(100)]):
        score = f1_score(y_true=y_true, y_pred=y_proba > threshold)
        if score > best_score:
            best_threshold = threshold
            best_score = score
    search_result = {'threshold': best_threshold, 'f1': best_score}
    return search_result
```

```python
th_sr = threshold_search(testy, test_preds)
print("Threshold Search F1 Result :")
print(th_sr)
```

Based on a probability threshold of 0.42, we can achieve an F1 score of 0.668. 

```python
def prob_to_bool(x, threshold=0.5):
    #function for thesholding predictions if less than 0.5
    x_bool = x>=threshold
    return x_bool
```

```python
final_pred = prob_to_bool(test_preds, threshold=0.42)
```

```python
print(classification_report(final_pred, testy))
```

```python
conf_matrix = confusion_matrix(testy, final_pred)

plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, xticklabels=["No Churn","Churn"], yticklabels=["No Churn","Churn"], 
            cmap='coolwarm', annot=True, fmt="d", annot_kws={"size": 20});
plt.title("Confusion Matrix - Light GBM")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

### Deciling and Lift

In some cases it is important to know where you stand with regards to targets you should pursue. This is where deciling and calculating lift comes into play. This is more of an area in targeting leads for sales but theres no reason we cant take a look at our top potential churn cases by lift.

```python
def decile_df(preds, actuals, y_train):
    deciles = pd.qcut(preds,q=10,labels=False)
    preds_df = pd.DataFrame(preds, index=actuals.index, columns=['preds'])
    preds_df['actuals'] = actuals.values
    preds_df['decile'] = deciles
    dec_df = preds_df.groupby('decile').agg(['min','max','mean','count'])
    dec_df.index = range(10,0,-1)
    dec_df.columns = ['_'.join(col).strip() for col in dec_df.columns.values]
    dec_df2 = dec_df[['preds_min','preds_max','actuals_mean','preds_count']]
    dec_df2.columns = ['min','max','p','n']
    baseline = (y_train.sum()/len(y_train))#[0]
    dec_df2['lift'] = dec_df2['p']/baseline
    return dec_df2.sort_index(), preds_df
```

```python
lr_dec, lr_dec_by_idx = decile_df(test_preds, testy, trainy)
```

```python
lr_dec
```

From our deciling we see that the top **4** deciles lift are > 1.0 so its these potential customers we would focus on first.

The top 106 customers in the first decile have a lift of 3.03 which means that based on their probability they are three times as likely to churn.

Here **p** is the mean of the actuals.

```python
fig, ax=plt.subplots(1,1,figsize=(12,8))
lr_dec[['min','max']].plot(kind='bar',ax=ax, alpha=0.5, rot=15);
plt.title('Decile Distributions');
```
